```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Generate synthetic data for demonstration purposes
np.random.seed(42)

# Simulating genetic variant data (instrumental variable)
n_samples = 1000
genetic_data = np.random.normal(0, 1, n_samples)

# Simulating exposure data (e.g., cholesterol levels)
exposure_data = np.random.normal(0, 1, n_samples)

# Simulating outcome data (e.g., heart disease)
# Outcome = Beta * Exposure + Error
beta = 0.5
error = np.random.normal(0, 1, n_samples)
outcome_data = beta * exposure_data + error

# Performing Mendelian randomization analysis
mr_results = sm.OLS(outcome_data, sm.add_constant(exposure_data)).fit()

# Printing summary of Mendelian randomization results
print(mr_results.summary())

# Visualization
plt.scatter(exposure_data, outcome_data, alpha=0.5)
plt.xlabel('Exposure (e.g., Cholesterol Levels)')
plt.ylabel('Outcome (e.g., Heart Disease)')
plt.title('Mendelian Randomization Analysis')
plt.grid(True)
plt.show()
```
